```python
colnames=["Sepal_length","Sepal_width","Petal_length","Petal_width","Class_labels"]
```

```python
df=pd.read_csv("iris.data",names=colnames)
df.head()
```

```python
df1=df.values
df1
```

```python
X=df1[:,0:4] #feature varibales
Y=df1[:,4] #target variables
print(Y)
```

```python
sb.pairplot(df,hue="Class_labels")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
model1=LogisticRegression()
```

```python
model1.fit(X_train,Y_train)
```

```python
model1.score(X_test,Y_test)
```

```python
y_predicted=model1.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_predicted,Y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_predicted))
```
